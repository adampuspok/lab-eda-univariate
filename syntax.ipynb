```python
import pandas as pd

df = pd.read_csv('amz_uk_price_prediction_dataset.csv')
```

```python
df.head()
```

```python
#PART1
#Task1

#TOP5: Sports & Outdoors, Beauty, Handmade Clothing, Shoes & Accessories, Bath & Body, Birthday Gifts 

df['category'].value_counts()
```

```python
#Task2

import matplotlib.pyplot as plt
import seaborn as sns

# Count categories
category_counts = df['category'].value_counts()

# Plot bar chart (top 10 for readability)
plt.figure(figsize=(12, 6))
sns.barplot(x=category_counts.head(10).values, y=category_counts.head(10).index, palette="viridis")
plt.title("Top 10 Product Categories by Frequency")
plt.xlabel("Number of Products")
plt.ylabel("Category")
plt.tight_layout()
plt.show()
```

```python
top_5 = category_counts.head(5)

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5.values, labels=top_5.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Proportion of Top 5 Product Categories")
plt.axis('equal')  # Equal aspect ratio ensures the pie is circular.
plt.show()
```

```python
#PART2
#Task1

prices = df['price'].dropna()

mean_price = prices.mean()

median_price = prices.median()

mode_price = prices.mode()

print(f"Mean price: {mean_price:.2f}")
print(f"Median price: {median_price:.2f}")
print(f"Mode price: {mode_price.values}")
```

```python
#Task2

variance = prices.var()

std_dev = prices.std()

price_range = prices.max() - prices.min()

Q1 = prices.quantile(0.25)
Q3 = prices.quantile(0.75)
iqr = Q3 - Q1

print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Range: {price_range:.2f}")
print(f"Interquartile Range: {iqr:.2f}")
```

```python
#Task3

import matplotlib.pyplot as plt
import seaborn as sns

# Drop missing prices
prices = df['price'].dropna()

# Histogram
plt.figure(figsize=(10, 5))
sns.histplot(prices, bins=50, kde=True, color='skyblue')
plt.title("Distribution of Product Prices")
plt.xlabel("Price (£)")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(prices[prices < 100], bins=50, kde=True, color='orange')
plt.title("Distribution of Product Prices (Under £100)")
plt.xlabel("Price (£)")
plt.ylabel("Number of Products")
plt.show()
```

```python
plt.figure(figsize=(10, 1.5))
sns.boxplot(x=prices, color='lightblue')
plt.xlim(0, 100)  # adjust this as needed
plt.title("Boxplot of Product Prices")
plt.xlabel("Price (£)")
plt.show()
```

```python
print(df['stars'].dtype)
print(df['stars'].unique())

df['stars'] = pd.to_numeric(df['stars'], errors='coerce')
```

```python
#PART3
#Task1

ratings_nonzero = df.loc[df['stars'] > 0, 'stars']

mean_rating = ratings_nonzero.mean()
median_rating = ratings_nonzero.median()
mode_rating = ratings_nonzero.mode()

print(f"Mean rating (excluding zeros): {mean_rating:.2f}")
print(f"Median rating (excluding zeros): {median_rating:.2f}")
print(f"Mode rating(s) (excluding zeros): {mode_rating.values}")
```

```python
#Task2

ratings = df.loc[df['stars'] > 0, 'stars'] 

variance = ratings.var()
std_dev = ratings.std()
Q1 = ratings.quantile(0.25)
Q3 = ratings.quantile(0.75)
iqr = Q3 - Q1

print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Interquartile Range (IQR): {iqr:.2f}")
```

```python
#Task3

ratings = df.loc[df['stars'] > 0, 'stars']

from scipy.stats import skew, kurtosis

rating_skewness = skew(ratings)
rating_kurtosis = kurtosis(ratings)

print(f"Skewness: {rating_skewness:.2f}")
print(f"Kurtosis: {rating_kurtosis:.2f}")
```

```python
#Task4

import matplotlib.pyplot as plt
import seaborn as sns

ratings = df.loc[df['stars'] > 0, 'stars']

plt.figure(figsize=(8,5))
sns.histplot(ratings, bins=5, discrete=True, color='dodgerblue')
plt.title("Distribution of Product Ratings (Stars)")
plt.xlabel("Rating (Stars)")
plt.ylabel("Number of Products")
plt.xticks([1, 2, 3, 4, 5])  # Assuming 1 to 5 star ratings
plt.show()
```
